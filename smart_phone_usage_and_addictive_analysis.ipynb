```python
 #Univariate analysis
 num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Count of {col}')
    plt.show()
```

```python
#Outlier Detection
#box plot method
for col in num_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#IQR method
def detect_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df[(df[col] < lower) | (df[col] > upper)]

for col in num_cols:
    print(col, len(detect_outliers(col)))
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#Numerical vs Numerical
sns.scatterplot(x='daily_screen_time_hours', y='addicted_label', data=df)
plt.show()
```

```python
#Categorical vs Numerical
sns.boxplot(x='gender', y='daily_screen_time_hours', data=df)
plt.show()
```

```python
#Categorical vs Categorical
pd.crosstab(df['gender'], df['addicted_label']).plot(kind='bar', stacked=True)
plt.show()
```

```python
#Feature engineering
# Feature 1
df['screen_time_level'] = pd.cut(
    df['daily_screen_time_hours'],
    bins=[0,3,6,10,df['daily_screen_time_hours'].max()],
    labels=['Low','Moderate','High','Extreme']
)

# Feature 2
df['usage_intensity'] = df['daily_screen_time_hours'] / df['daily_screen_time_hours'].max()

# Feature 3
df['addicted'] = df['addicted_label'].apply(lambda x: 1 if x == 'Yes' else 0)

# SHOW OUTPUT
df[['daily_screen_time_hours','screen_time_level','usage_intensity','addicted']].head()
```
